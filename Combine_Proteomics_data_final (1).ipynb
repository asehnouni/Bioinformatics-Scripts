```python
unique_sources = pd.concat([output1['Source'], output2['Source']]).unique()

# Create a DataFrame to hold the merged data
merged_output = pd.DataFrame(columns=output1.columns)

# Iterate through unique sources and merge the rows
for source in unique_sources:
    source_data1 = output1[output1['Source'] == source]
    source_data2 = output2[output2['Source'] == source]

    # Merge data based on the index (common columns)
    combined_source_data = pd.concat([source_data1, source_data2], axis=1)
    
    # Remove duplicated "Source" columns
    combined_source_data = combined_source_data.loc[:,~combined_source_data.columns.duplicated()]

    # Add 'Source' column from output2 to combined_source_data
    combined_source_data['Source'] = source

    merged_output = pd.concat([merged_output, combined_source_data], ignore_index=True)
```
